# RMSD PLOT

```python
import numpy as np
import matplotlib.pyplot as plt

file_path_3ql3 ='./3ql3_rmsd_30ns.agr' 
file_path_3ql0 ='./3ql0_rmsd_30ns.agr'
file_path_4m6k = './4m6k_rmsd.agr' 


data_3ql3 = np.loadtxt(file_path_3ql3, skiprows=8)
data_3ql0 = np.loadtxt(file_path_3ql0, skiprows=8)
data_4m6k = np.loadtxt(file_path_4m6k, skiprows=8)


time_3ql3 = data_3ql3[:, 0] / 200 
rmsd_3ql3 = data_3ql3[:, 1]       

time_3ql0 = data_3ql0[:, 0] / 200  
rmsd_3ql0 = data_3ql0[:, 1]        

time_4m6k = data_4m6k[:, 0] / 200  
rmsd_4m6k = data_4m6k[:, 1]   

plt.figure(figsize=(8, 6))

plt.plot(time_3ql3, rmsd_3ql3, color='green', linestyle='-', label='3QL3')
plt.plot(time_3ql0, rmsd_3ql0, color='blue', linestyle='-', label='3QL0')
plt.plot(time_4m6k, rmsd_4m6k, color='purple', linestyle='-', label='4M6K')

plt.xlim(-0.1,30)
plt.ylim(0.5,3.5)

plt.xlabel('Time (nsec)')
plt.ylabel('RMSD ()')
#plt.title('RMSD over Time')
plt.legend()
#plt.grid(True)
plt.tight_layout()


#plt.savefig('rmsd_30ns_plot_ec-hDHFR.jpg', dpi=300) 
plt.show()

#calculate average

import statistics
#print(time_3ql3)
average_3ql3=statistics.mean(rmsd_3ql3)
print('3ql3 dist ave: ',average_3ql3)

average_3ql0=statistics.mean(rmsd_3ql0)
print('3ql0 dist ave: ',average_3ql0)

average_4m6k=statistics.mean(rmsd_4m6k)
print('4m6k dist ave: ',average_4m6k)
```
